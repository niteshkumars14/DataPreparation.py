```python
# Importing pandas Library.
import os
import pandas as pd
import numpy as np
```

```python
# importing csv.
df = pd.read_csv('data.csv')
```

```python
# Check Number of columns & Row.
df.shape
```

```python
# Check Column Names.
df.columns
```

```python
# Check Datatypes
df.dtypes
```

```python
# Check first 5 Rows of the Data.
df.head()
```

```python
# Check Last 5 Rows of the Data.
df.head()
```

```python
# Remove Rows where all columns are blank.
df = df.dropna(axis=0, how='all').reset_index(drop=True)
```

```python
df.head(4)
```

```python
# Check if the data has duplicates.
# If yes, Remove it.
df = df.drop_duplicates()
```

```python
# Check missing Values of each Column.
df.isnull().sum()
```

```python
# Fill Missing Values.
# Missing Values should be filled with a value which satisfy both business & statistical aspects.
# Replacing Null Values with mean.
df['Daily Charges MV'].fillna(df['Daily Charges MV'].mean(), inplace=True)
```

```python
# Check missing Values of DataFrame.
df.isnull().sum().sum()
```

```python
# Changing Column Datatypes.
# Initially Python read "Area code" columns as "FLOAT" datatype because it contains only Numbers.
# But "Area Code" belongs to a Categorical Column.
# Converting "Area Code" to class 'Object'
df['Area Code'] = df['Area Code'].astype('str')
df['Area Code'].dtype
```

```python
# Check DataFrame Summary.
df.describe()
```

```python
# Check Correlation.
from matplotlib import pyplot as plt
import seaborn as sns
corr = df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="Greens",annot=True)
```

### Correlation Matrix can be confusing sometimes, specially when you have more columns.
### There is another of checking correlation when you have more columns.

```python
corr[corr < 1].unstack().transpose()\
    .sort_values( ascending=False)\
    .drop_duplicates()
```
